## 구구단 출력

출력 결과 예시
```
----- 2 단 -----
2 * 1 = 2
2 * 2 = 4
2 * 3 = 6
...

----- 3단 -----
3 * 1 = 3
3 * 2 = 6
3 * 3 = 9
...

----- 9단 -----
...
9 * 8 = 72
9 * 9 = 81
```

````python
# GO!

for i in range(2, 10):  # 2 ~ 9
    print(f'\n-------------{i}단---------------')
    for k in range(1, 10):  # 1 ~ 9
        result = i * k
        print(f'{i} * {k} = {result}')
        
````

## 소수 찾기

> 2 부터 N 까지의 정수에서 소수를 찾아내기.

출력 결과 예시(N 이 10 일 경우)
```
2
3
5
7
```

### 추가 구현
> 소수 판단 과정에서 소수가 아니라 판단되면, `break` 사용

````python
# GO!
li = []
n = int(input('자연수를 적으세요.:  '))
for i in range(1, n + 1):
    for j in range(1, n + 1):
        if i % j == 0 and int(i / j) != 1 and int(i / j) != i:
            li.append(i)

result = set(range(1, n + 1)) - set(li)
print(result)
````

````python
##2 

N = n

for i in range(2, N + 1):  # 2 ~ 10
    

# i를 1부터 i까지의 숫자로 나눠보고, 1과 i로만 나누었을 때 나머지가 0이라면

    is_prime = True        # is로 시작하는 데이터는 무조건 T or F
    for j in range(2, i):  # 나랑 1은 나누면 무조건 0이니까 나눠지면 이제 무조건 소수
        if i % j == 0:
            is_prime = False
            break

    else:
        print(i)   # N + 1 까지 볼 필요 없이 제곱근 전까지만 보면 된다. 제곱근보다 큰 수끼리 곱해서 나올 수 없기 떄문이다.
````

## 팩토리얼 구현하기

> n! = 1 * 2 * 3 ... n-1 * n

출력 결과 예시
```
N 을 입력하세요: 5
120
```

````python
# GO!
result = 1
num = int(input('자연수를 입력하세요.:  '))

# 3 = 3 * 2 * 1
for i in range(1, num + 1):  # 1, 2, 3
    result = result * i
    
print(result)
````

````python
import math

result_2 = math.factorial(num)
print(result_2)
````

## 중첩 조건문 활용하기

> 아래와 같이 유저 목록이 담긴 리스트가 있습니다. 
```
users = [
    {'id': 'john', 'password': 'qwer1234'},
    {'id': 'neo', 'password': '12341234'},
    {'id': 'jason', 'password': 'kingjason'},
]
```
1) 사용자의 이름을 입력받아 
2) 등록된 사용자일 경우 
3) 비밀번호를 입력 받습니다. 
- 이때 비밀번호가 12345678의 경우 '환영합니다!' 라는 메시지를 출력하고 비밀번호가 틀릴 경우 '패스워드가 올바르지 않습니다.' 라는 메시지를 출력합니다. 
- 등록된 사용자가 아닌 경우 '존재하지 않는 사용자입니다.' 라는 메시지를 출력합니다. 

출력 결과 예시

```
id 를 입력하세요: neo
password 를 입력하세요: 12341234
환영합니다!
```

```
id 를 입력하세요: john
password 를 입력하세요: 123123
패스워드가 올바르지 않습니다.
```

```
id 를 입력하세요: JASON
존재하지 않는 사용자 입니다.
```

### 추가 구현
> id 혹은 pw 가 틀렸을 경우 맞을 때 까지 계속 물어보게 구현하기

````python
# users 안의 id 와 password 는 바꿔도 상관 없습니다.
users = [
    {'id': 'john', 'password': 'qwer1234'},
    {'id': 'neo', 'password': '12341234'},
    {'id': 'jason', 'password': 'kingjason'},
]

    
id_1 = input('id를 입력하세요:  ')

for i in range(len(users)):
    
    if id_1 != users[i]['id']:
        print('존재하지 않는 사용자입니다.')
        break
    else:
        pw_1 = input('password 를 입력하세요:  ')

        if users[i]['password'] == pw_1:
            print('환영합니다!')
            break
        else:
            print('패스워드가 올바르지 않습니다.')
            break
````

## 과일은 몇개?
아래 코드에서 과일과 아닌 것을 구분하기
basket = {'apple':4, 'orange':3, 'computer':28,
'tv':2, 'banana':5, 'tomato':9
}

fruits = ['apple', 'orange', 'banana', 'pear', 'tomato']

과일은 21개, 아닌 것은 30개


````python
basket = {'apple':4, 'orange':3, 'computer':28, 'tv':2, 'banana':5, 'tomato':9 }
fruits = ['apple', 'orange', 'banana', 'pear', 'tomato']
key_lis = list(basket.keys())
furits_num = 0
no_num = 0

for fruit in fruits:
    for key_li in key_lis:
        if fruit == key_li:
            furits_num += basket[key_li]
for value in basket.values():
        no_num += value

result_num = no_num - furits_num

print(f'과일은 {furits_num}개, 아닌 것은 {result_num}개')
````
